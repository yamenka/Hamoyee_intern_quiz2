```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge , Lasso
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
data.drop(columns=['date','lights'],inplace=True)
```

```python
data.head()
```

```python
scaler = MinMaxScaler()
normalised = pd.DataFrame(scaler.fit_transform(data),columns = data.columns)
```

```python
normalised.head()
```

```python
used_data = normalised.drop(columns='Appliances')
target = normalised['Appliances']
```

```python
used_data.head()
```

```python
target.head()
```

```python
X_train , X_test , Y_train, Y_test = train_test_split(used_data, target, test_size = 0.3, random_state = 42)
```

```python
linear_reg_model = LinearRegression()
```

```python
linear_reg_model.fit(X_train, Y_train)
```

```python
model_predictions = linear_reg_model.predict(X_test)
```

```python
mae = mean_absolute_error(Y_test, model_predictions)

rss = np.sum(np.square(Y_test - model_predictions))


mse = mean_squared_error(Y_test, model_predictions)


r2 = r2_score(Y_test , model_predictions)

```

```python

print(f'mae = {mae:.2f}\nrss = {rss:.2f}\nmse = {mse:.2f}\nr2_score = {r2:.2f}')
```

```python
x = normalised['T2']
y = normalised['T6']
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state= 42)
```

```python
model = LinearRegression()
model.fit(x_train.values.reshape(-1,1), y_train.values.reshape(-1,1))
```

```python
r2 = model.score(x_test.values.reshape(-1,1), y_test.values.reshape(-1,1))
```

```python
r2
```

```python
def get_weights(model, predictors, name):
    weights = pd.Series(model.coef_, predictors.columns).sort_values()
    weight_df = pd.DataFrame(weights).reset_index()
    weight_df.columns = ['features', name]
    weight_df[name].round(3)
    return weight_df
```

```python
weights = get_weights(linear_reg_model, X_train, 'Linear regression' )
```

```python
weights
```

```python
ridge = Ridge(alpha=0.4).fit(X_train, Y_train)
```

```python
ridge_pred = ridge.predict(X_test)
```

```python
ridge_mae = mean_absolute_error(Y_test, ridge_pred)

ridge_rss = np.sum(np.square(Y_test - ridge_pred))

ridge_mse = mean_squared_error(Y_test, ridge_pred)

ridge_r2 = r2_score(Y_test, ridge_pred)
```

```python
print(f'mae = {ridge_mae}\nrss = {ridge_rss}\nmse = {ridge_mse}\nr2_score = {ridge_r2}')
```

```python
ridge_weights = get_weights(ridge, X_train, "Ridge Model")
```

```python
ridge_weights
```

```python
lasso_model = Lasso(alpha=0.001).fit(X_train,Y_train)
lasso_pred = lasso_model.predict(X_test)

```

```python
#Mean absolute error
lasso_mae = mean_absolute_error(Y_test, lasso_pred)

#Residual sum of squares 
lasso_rss = np.sum(np.square(Y_test - lasso_pred))

#Mean squared error
lasso_mse = mean_squared_error(Y_test, lasso_pred)

#R squared score
lasso_r2 = r2_score(Y_test, lasso_pred)
```

```python
print(f'mae = {lasso_mae}\nrss = {lasso_rss}\nmse = {lasso_mse}\nr2_score = {lasso_r2}')
```

```python
lasso_weights = get_weights(lasso_model, X_train, 'Lasso Model')
lasso_weights[lasso_weights['Lasso Model'] != 0]
```

```python

final_weights = weights.merge(ridge_weights, on='features')
final_weights = final_weights.merge(lasso_weights, on='features')
```
